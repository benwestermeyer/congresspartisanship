```python
import requests
import shutil

import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning) # dealing with warnings. is this okay?

def callme(congress, bill):
    url = "https://www.govtrack.us/congress/votes/114-2016" # we are looking at bills from 2016, the 114th congress
    year = url[-4:]
    url = url + "/" + congress
    url = url + bill
    url = url + "/export/csv"
    print url
    r = requests.get(url, verify=False, stream=True, timeout=10.0)
    if r.status_code != 200:
        print "Failure!"
        exit()
    else:
        r.raw.decode_content = True
        with open("%s%s%s.csv" % (year, congress, bill), 'wb') as f:
            shutil.copyfileobj(r.raw, f)
        print "Success"

for i in range(1,164): # the reason that we loop through 163 is that there were 163 senate bills in 2016
    callme("s", str(i))
```
